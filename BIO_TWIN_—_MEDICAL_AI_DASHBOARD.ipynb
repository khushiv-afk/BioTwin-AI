===============================
🧬 BioTwin: Clinical AI-Based Digital Health Twin System
===============================

📌 PROJECT OVERVIEW:
BioTwin is a clinical AI-inspired system that simulates a digital health twin. It predicts health risks using basic lifestyle parameters like sleep, stress, age, and physical activity.

The system converts user inputs into a structured medical-style report with risk classification.

---

📌 PROBLEM STATEMENT:
Modern healthcare lacks continuous real-time monitoring of lifestyle-based health risks. Most people are unaware of early indicators of diseases such as diabetes and cardiac issues.

BioTwin addresses this by providing a simple AI-based simulation that estimates health risks using lifestyle data.

---

📌 OBJECTIVE:
To develop an AI-based system that:
- Collects lifestyle inputs
- Computes health risk score
- Classifies disease risk levels
- Generates a medical-style report for early awareness

---

📌 INPUT PARAMETERS:
The system takes the following inputs:
- Age of user
- Daily sleep hours
- Daily step count
- Stress level (0–10 scale)

---

📌 METHODOLOGY / WORKING:

Step 1: Collect user lifestyle data  
Step 2: Apply rule-based risk scoring model  
Step 3: Compute overall health score (0–100 scale)  
Step 4: Estimate disease risks:
        - Diabetes Risk
        - Cardiac Risk  
Step 5: Classify risk levels:
        - Low Risk
        - Moderate Risk
        - High Risk  
Step 6: Display clinical-style dashboard report

---

📌 RISK MODEL (LOGIC):
Health risk is calculated using weighted lifestyle factors:

- Poor sleep increases risk
- High stress increases risk
- Low physical activity increases risk

Final output is normalized into:
- Health Score (0–100)
- Disease Risk Percentages

---

📌 OUTPUTS GENERATED:
The system produces:
- Overall Health Score
- Diabetes Risk Percentage
- Cardiac Risk Percentage
- Clinical Insight Message

---

📌 KEY FEATURES:
- Interactive medical AI dashboard
- Real-time risk prediction
- Color-coded severity classification
- Explainable rule-based AI model
- Simple and interpretable clinical logic

---

📌 CONCLUSION:
BioTwin demonstrates how AI-inspired rule-based systems can simulate early health risk detection and provide a structured health report similar to clinical decision support systems.

It serves as a prototype for future AI-based healthcare monitoring systems.

```python
import ipywidgets as widgets
from IPython.display import display, HTML

# ---------------- UI STYLE (MEDICAL DASHBOARD LOOK) ----------------
display(HTML("""
<style>
    .title {
        font-size: 28px;
        font-weight: bold;
        color: #0b3d91;
        margin-bottom: 10px;
    }
    .card {
        padding: 15px;
        border-radius: 12px;
        background-color: #f5f9ff;
        border: 1px solid #d6e4ff;
        margin-top: 10px;
        font-family: Arial;
    }
    .risk-high { color: red; font-weight: bold; }
    .risk-mid { color: orange; font-weight: bold; }
    .risk-low { color: green; font-weight: bold; }
</style>
"""))

display(HTML("<div class='title'>🧬 BioTwin Clinical AI Dashboard</div>"))

# ---------------- INPUTS ----------------
age = widgets.IntSlider(value=30, min=18, max=80, description='Age')
sleep = widgets.IntSlider(value=6, min=0, max=12, description='Sleep')
steps = widgets.IntSlider(value=6000, min=0, max=20000, description='Steps')
stress = widgets.IntSlider(value=5, min=0, max=10, description='Stress')

button = widgets.Button(description="Run Clinical Analysis", button_style='primary')

output = widgets.Output()

# ---------------- MODEL ----------------
def run_model(b):
    with output:
        output.clear_output()

        # Risk engine (simple clinical scoring)
        risk_score = (stress.value * 12) + (8 - sleep.value) * 10 + (5000 - min(steps.value,5000))/400

        health_score = max(0, 100 - risk_score)
        diabetes_risk = min(100, risk_score + 12)
        cardiac_risk = min(100, risk_score)

        # ---------------- RISK LEVEL ----------------
        def level(x):
            if x < 40:
                return "LOW RISK", "risk-low"
            elif x < 70:
                return "MODERATE RISK", "risk-mid"
            else:
                return "HIGH RISK", "risk-high"

        d_level, d_class = level(diabetes_risk)
        c_level, c_class = level(cardiac_risk)

        # ---------------- DASHBOARD UI ----------------
        display(HTML(f"""
        <div class='card'>
            <h3>📊 Patient Summary Report</h3>
            <p><b>Age:</b> {age.value}</p>
            <p><b>Health Score:</b> {health_score:.2f} / 100</p>
        </div>

        <div class='card'>
            <h3>🩸 Diabetes Risk</h3>
            <p class='{d_class}'>{d_level} ({diabetes_risk:.2f}%)</p>
        </div>

        <div class='card'>
            <h3>❤️ Cardiac Risk</h3>
            <p class='{c_class}'>{c_level} ({cardiac_risk:.2f}%)</p>
        </div>

        <div class='card'>
            <h3>📌 Clinical Insight</h3>
            <p>{
                "Patient shows stable vitals." if health_score > 70 else
                "Lifestyle intervention recommended." if health_score > 40 else
                "Immediate clinical attention suggested."
            }</p>
        </div>
        """))

button.on_click(run_model)

# ---------------- DISPLAY ----------------
display(age, sleep, steps, stress, button, output)
```
